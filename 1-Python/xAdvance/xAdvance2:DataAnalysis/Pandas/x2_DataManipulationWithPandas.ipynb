```python
import pandas as pd

data = pd.read_csv('x1.csv')

# Hanling missing values
data.isnull() # if a statement is true, it means that the value is missing in that cell
data.isnull().sum() # Count missing values in each column
data[data.isnull().any(axis=1)] # axis = 1 - rows and axis = 0 - columns
data.dropna() # Drop rows with missing values
data.dropna(axis=1) # Drop columns with missing values
data.dropna(thresh=2) # Drop rows with 2 or more missing values
data.fillna(0) # Fill missing values with 0
# Renaming columns
data.rename(columns = {'Date': 'Tareek', 'Unit Price': 'Aukaat'}, inplace = True)
data.head(5)
# Change Data types
data['RajatKeOrders'] = data["Transaction ID"].astype('string')
data.head(5)

# ChatGPT can help you with all this
```

```python
# Merging and Joinging DataFrames

df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})

# print(df1)
# print(df2)

# Inner Join
df3 = pd.merge(df1, df2, on = 'Key', how = 'inner')
# print(df3)

# Left Join : All rows from df1 and only matching rows from df2
df4 = pd.merge(df1, df2, on = 'Key', how = 'left')
# print(df4)

# Right Join : All rows from df2 and only matching rows from df1
df5 = pd.merge(df1, df2, on = 'Key', how = 'right')
print(df5)
```
