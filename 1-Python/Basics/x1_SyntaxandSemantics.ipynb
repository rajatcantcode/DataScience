In Python, indentation is used to define the structure of the code, such as blocks for loops, functions, and conditionals. Here are the key points:

Indentation Level: Use 4 spaces for each level of indentation.
Consistent Indentation: Always use the same number of spaces. Do not mix spaces and tabs.
Code Blocks: Indent code inside functions, loops, and conditionals.

```python

if 5>2:
    # Indent code inside this block
    print("Condition is true")
else:
    # Indent code inside this block
    print("Condition is false")

#No Indentation: Top-level code starts at the leftmost position.
#Indented Code: Each nested block increases indentation by 4 spaces.
#Simple rule: Use 4 spaces to indent code blocks consistently.'''
```
